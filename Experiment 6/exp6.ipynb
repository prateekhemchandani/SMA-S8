```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("engagement_dataset.csv")
```

```python
print(df.head())
```

```python
# remove nulls
df = df.dropna()
# remove duplicates
df = df.drop_duplicates()
```

```python
df['engagement'] = df['likes'] + df['comments'] + df['shares']
```

```python
content_engagement = df.groupby('content_type')['engagement'].mean()
print(content_engagement)
```

```python
user_engagement = df.groupby('user_type')['engagement'].mean()
print(user_engagement)
```

```python
top_post = df.loc[df['engagement'].idxmax()]
print(top_post)
```

Content type Engagement (visualisation)

```python
plt.hist(df['engagement'])
plt.xlabel("Engagement")
plt.ylabel("Frequency")
plt.title("Distribution of Engagement")
plt.show()
```

User engagement (visualisation)

```python
plt.figure()
user_engagement.plot(kind='bar')

plt.title("Engagement by User Type")
plt.xlabel("User Type")
plt.ylabel("Average Engagement")

plt.show()
```
