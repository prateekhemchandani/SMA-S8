```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("trend_analysis_dataset.csv")   
```

```python
df['timestamp']
```

```python
# remove nulls
df = df.dropna(subset=['timestamp'])

# remove duplicates
df = df.drop_duplicates()

# convert to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

**Group by Date**

```python
# extract only date
df['date'] = df['timestamp'].dt.date

# count posts per day
trend = df.groupby('date').size()

print(trend.head())
```

```python
plt.figure()
trend.plot(kind='line')

plt.xlabel("Date")
plt.ylabel("Number of Posts")
plt.title("Trend Analysis of Social Media Activity")

plt.xticks(rotation=45)
plt.show()
```
